```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA
```

## Exercise 1

```python
# item 1

N_SAMPLES = 100

CLASS_0 = ( (2, 3), (0.8, 2.5))
CLASS_1 = ( (5, 6), (1.2, 1.9))
CLASS_2 = ( (8, 1), (0.9, 0.9))
CLASS_3 = ((15, 4), (0.5, 2.0))

def generate_data(CLASS_data, n_samples) -> np.ndarray[np.ndarray, np.ndarray]:
    dim_0 = np.random.normal(loc=CLASS_data[0][0], scale=CLASS_data[1][0], size=n_samples)
    dim_1 = np.random.normal(loc=CLASS_data[0][1], scale=CLASS_data[1][1], size=n_samples)
    return np.array(list(zip(dim_0, dim_1)))

class_0_data = generate_data(CLASS_0, N_SAMPLES)
class_1_data = generate_data(CLASS_1, N_SAMPLES)
class_2_data = generate_data(CLASS_2, N_SAMPLES)
class_3_data = generate_data(CLASS_3, N_SAMPLES)
```

```python
# item 2

plt.scatter(class_0_data[:,0], class_0_data[:,1], color='red', label='Classe 0', alpha=0.1)
plt.scatter(class_1_data[:,0], class_1_data[:,1], color='blue', label='Classe 1', alpha=0.1)
plt.scatter(class_2_data[:,0], class_2_data[:,1], color='green', label='Classe 2', alpha=0.1)
plt.scatter(class_3_data[:,0], class_3_data[:,1], color='purple', label='Classe 3', alpha=0.1)
plt.xlabel('Dimenso 0')
plt.ylabel('Dimenso 1')
plt.title('Dados Gerados')
plt.legend()
plt.show()
```

### Item 3

**a.** Observing the scatter plot, a few things stand out.
- Class 3 is very easily distinguishable from the others, since its values of X (dim 0) are fairly distant from all the others'.
- Class 2 is much closer to the others, but there is pretty much no overlap with class 0 and a possibly a little with class 1.
- Classes 0 and 1 have considerable visual overlap.

**b.** One single linear boundary, absolutely not. Three linear boundaries, yes.
