````python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA
````

## Exercise 1

````python
# item 1

N_SAMPLES = 100

CLASS_0 = ( (2, 3), (0.8, 2.5))
CLASS_1 = ( (5, 6), (1.2, 1.9))
CLASS_2 = ( (8, 1), (0.9, 0.9))
CLASS_3 = ((15, 4), (0.5, 2.0))

def generate_data(CLASS_data, n_samples) -> np.ndarray[np.ndarray, np.ndarray]:
    dim_0 = np.random.normal(loc=CLASS_data[0][0], scale=CLASS_data[1][0], size=n_samples)
    dim_1 = np.random.normal(loc=CLASS_data[0][1], scale=CLASS_data[1][1], size=n_samples)
    return np.array(list(zip(dim_0, dim_1)))

class_0_data = generate_data(CLASS_0, N_SAMPLES)
class_1_data = generate_data(CLASS_1, N_SAMPLES)
class_2_data = generate_data(CLASS_2, N_SAMPLES)
class_3_data = generate_data(CLASS_3, N_SAMPLES)
````

````python
# item 2

plt.scatter(class_0_data[:,0], class_0_data[:,1], color='red', label='Classe 0', alpha=0.1)
plt.scatter(class_1_data[:,0], class_1_data[:,1], color='blue', label='Classe 1', alpha=0.1)
plt.scatter(class_2_data[:,0], class_2_data[:,1], color='green', label='Classe 2', alpha=0.1)
plt.scatter(class_3_data[:,0], class_3_data[:,1], color='purple', label='Classe 3', alpha=0.1)
plt.xlabel('Dimensão 0')
plt.ylabel('Dimensão 1')
plt.title('Dados Gerados')
plt.legend()
plt.show()
````

### Item 3

**a.** Observing the scatter plot, a few things stand out.
- Class 3 is very easily distinguishable from the others, since its values of X (dim 0) are fairly distant from all the others'.
- Class 2 is much closer to the others, but there is pretty much no overlap with class 0 and a possibly a little with class 1.
- Classes 0 and 1 have considerable visual overlap.

**b.** One single linear boundary, absolutely not. Three linear boundaries, yes.

## Exercise 2

````python
# item 1

class_A_mean = np.zeros(5)
class_B_mean = np.array([1.5] * 5)

class_A_cov = np.array([
    [1,0.8,0.1,0,0],
    [0.8,1,0.3,0,0],
    [0.1,0.3,1,0.5,0],
    [0,0,0.5,1,0.2],
    [0,0,0,0.2,1]
])

class_B_cov = np.array([
    [1,-0.7,0.2,0,0],
    [-0.7,1.5,0.4,0,0],
    [0.2,0.4,1.5,0.6,0],
    [0,0,0.6,1.5,0.3],
    [0,0,0,0.3,1.5]
])

class_A_data = np.random.multivariate_normal(mean=class_A_mean, cov=class_A_cov, size=500)
class_B_data = np.random.multivariate_normal(mean=class_B_mean, cov=class_B_cov, size=500)

pca = PCA(n_components=2)
class_A_data_2d = pca.fit_transform(class_A_data)
class_B_data_2d = pca.transform(class_B_data)

plt.scatter(class_A_data_2d[:,0], class_A_data_2d[:,1], color='red', label='Classe A', alpha=0.1)
plt.scatter(class_B_data_2d[:,0], class_B_data_2d[:,1], color='blue', label='Classe B', alpha=0.1)
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('Dados Projetados em 2D via PCA')
plt.legend()
plt.show()
````

### Item 3

**a.** There seems to be a fair amount of overlap between the two classes.

**b.** Since there is a bit of overlap between the classes when reducing them to 2D, a linear separation of the two seems impossible.

### Item 3

- Missing Values:
    - Categorical: fill out the values randomly but keeping the same proportions of the values not missing (except ```Name``` and ```Cabin```)
    - Numerical: fill out with the average
    - Cabin: will be split into ```CabinSide``` and ```CabinDeck``` and then those two will be filled out similarly

````python
df_train["CabinSide"] = df_train["Cabin"].str[-1]
df_train["CabinDeck"] = df_train["Cabin"].str.split('/').str[0]
df_train = df_train.drop(columns=['Cabin'])
````

````python
categorical_columns = ['HomePlanet', 'CryoSleep', 'Destination', 'VIP', 'CabinSide', 'CabinDeck']

for column in categorical_columns:
    print(f"Proportions for {column}:")
    print(df_train[column].value_counts(normalize=True))
    print()
````

````python
for column in categorical_columns:
    proportions = df_train[column].value_counts(normalize=True, dropna=True)
    missing_count = df_train[column].isnull().sum()
    df_train.loc[df_train[column].isnull(), column] = np.random.choice(
        proportions.index, size=missing_count, p=proportions.values
    )
````

````python
for column in categorical_columns:
    print(f"Proportions for {column}:")
    print(df_train[column].value_counts(normalize=True))
    print()
````

````python
numerical_columns = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']

for column in numerical_columns:
    mean_value = df_train[column].mean()
    df_train[column].fillna(mean_value, inplace=True)
````

````python
missing_values = df_train.isnull().sum() / len(df_train)
missing_values
````

````python
# one-hot encode categorical variables with more than two categories
one_hot_encode_columns = ['HomePlanet', 'Destination', 'CabinDeck']
df_train = pd.get_dummies(df_train, columns=one_hot_encode_columns, drop_first=True)
print(df_train.head())
````

````python
# calculate z-scores for numerical columns
for column in numerical_columns:
    df_train[column] = (df_train[column] - df_train[column].mean()) / df_train[column].std()
````

````python
# change true and false to 1 and 0
df_train['CryoSleep'] = df_train['CryoSleep'].astype(int)
df_train['VIP'] = df_train['VIP'].astype(int)
df_train['CabinSide'] = df_train['CabinSide'].map({'P': 1, 'S': 0})
````
